1. What is hypothesis testing in statistics ?
>>>**a. Definition:** Hypothesis testing is a method used to determine if there is enough evidence to reject a null hypothesis (H₀) in favor of an alternative hypothesis (H₁).

**b. Steps:**
   - Formulate two hypotheses: the null hypothesis (H₀) and the alternative hypothesis (H₁).
   - Choose a significance level (α), usually 0.05.
   - Collect data and calculate the test statistic.
   - Compare the test statistic with the critical value or use the p-value to make a decision.
   - If the p-value < α, reject H₀. Otherwise, fail to reject H₀.

**c. Purpose:** Hypothesis testing helps assess the validity of assumptions or claims about a population based on sample data

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
>>> **a. Null Hypothesis (H₀):** No effect, difference, or relationship in the population.  
**b. Alternative Hypothesis (H₁):** There is an effect, difference, or relationship in the population.  
**c. Key Difference:** H₀ suggests no change, H₁ suggests a change.

3. What is the significance level in hypothesis testing, and why is it important?
>>> **a. Significance Level (α):** The probability of rejecting the null hypothesis when it is actually true, commonly set at 0.05.

**b. Importance:** It determines the threshold for deciding when to reject the null hypothesis. A lower α reduces the risk of a Type I error (false positive).

4. What does a P-value represent in hypothesis testing?
>>> **a. P-value:** The probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis is true.

**b. Interpretation:** A low p-value (typically < 0.05) indicates strong evidence against the null hypothesis, while a high p-value suggests weak evidence.

5.  How do you interpret the P-value in hypothesis testing?
>>> **a. Low P-value (≤ α):** Reject the null hypothesis (strong evidence against H₀).  
**b. High P-value (> α):** Fail to reject the null hypothesis (weak evidence against H₀).  
**c. Significance Level (α):** The threshold for deciding if the p-value is low enough to reject H₀, typically set at 0.05.

6. What are Type 1 and Type 2 errors in hypothesis testing?
>>> **a. Type 1 Error:** Rejecting the null hypothesis when it is actually true (false positive).

**b. Type 2 Error:** Failing to reject the null hypothesis when it is actually false (false negative).

**c. Trade-off:** Reducing one error increases the likelihood of the other.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
>>> **a. One-tailed Test:** Tests for an effect in one direction (either greater than or less than a certain value).

**b. Two-tailed Test:** Tests for an effect in both directions (either greater than or less than a certain value).

**c. Key Difference:** One-tailed tests focus on a specific direction, while two-tailed tests consider both possible directions.

8. What is the Z-test, and when is it used in hypothesis testing?
>>> **a. Z-test:** A statistical test used to determine if there is a significant difference between sample data and a population when the population standard deviation is known or the sample size is large.

**b. When to Use:** It is used when the sample size is large (n > 30) or when the population standard deviation is known, and the data is approximately normally distributed.

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?
>>> **a. Z-score Formula:**  
\[
Z = \frac{X - \mu}{\sigma / \sqrt{n}}
\]
Where:  
- X = sample mean  
- μ = population mean  
- σ = population standard deviation  
- n = sample size  

**b. Interpretation:** The Z-score represents how many standard deviations the sample mean is away from the population mean. A high absolute Z-score indicates a significant difference between the sample and population.

10.  What is the T-distribution, and when should it be used instead of the normal distribution?
>>> **a. T-distribution:** A probability distribution that is similar to the normal distribution but has heavier tails, which account for greater variability in smaller samples.

**b. When to Use:** It is used when the sample size is small (n ≤ 30) or the population standard deviation is unknown. The t-distribution becomes closer to the normal distribution as the sample size increases.

11. What is the difference between a Z-test and a T-test?
>>> **a. Z-test:** Used when the population standard deviation is known or the sample size is large (n > 30). Assumes normal distribution.

**b. T-test:** Used when the population standard deviation is unknown or the sample size is small (n ≤ 30). Uses the t-distribution.

**c. Key Difference:** Z-test is for large samples or known population SD, while T-test is for small samples or unknown population SD.

12. What is the T-test, and how is it used in hypothesis testing?
>>> **a. T-test:** A statistical test used to compare the means of two groups or a sample mean to a population mean when the sample size is small or the population standard deviation is unknown.

**b. How It’s Used:** It calculates the difference between the sample mean and the population mean (or between two sample means) in terms of standard error, then compares this to the t-distribution to determine if the difference is statistically significant.

13. What is the relationship between Z-test and T-test in hypothesis testing?
>>> **a. Z-test vs T-test:** Both tests are used to compare means, but the Z-test is for large samples or known population standard deviation, while the T-test is for small samples or unknown population standard deviation.

**b. Relationship:** As the sample size increases, the T-distribution approaches the normal distribution, making the T-test similar to the Z-test for large samples.

14. What is a confidence interval, and how is it used to interpret statistical results?
>>> **a. Confidence Interval:** A range of values, derived from sample data, that is likely to contain the population parameter (e.g., mean) with a specified level of confidence (e.g., 95%).

**b. Use in Interpretation:** It provides an estimate of the range within which the true population parameter lies. A wider interval suggests more uncertainty, while a narrower one indicates more precision.

15. What is the margin of error, and how does it affect the confidence interval?
>>> **a. Margin of Error:** The amount added and subtracted from the sample estimate to create the confidence interval, representing the range of uncertainty around the estimate.

**b. Effect on Confidence Interval:** A larger margin of error results in a wider confidence interval, indicating more uncertainty. A smaller margin of error narrows the interval, providing a more precise estimate.

16.  How is Bayes' Theorem used in statistics, and what is its significance?
>>> **a. Bayes' Theorem:** A mathematical formula used to update the probability of a hypothesis based on new evidence. It relates the conditional probability of events using prior knowledge and likelihood.

**b. Significance:** Bayes' Theorem allows for more accurate predictions and decision-making by incorporating prior information and adjusting probabilities as new data becomes available.

17. What is the Chi-square distribution, and when is it used?
>>> **a. Chi-square Distribution:** A probability distribution used to analyze categorical data, particularly for goodness-of-fit tests, tests of independence, and tests of variance.

**b. When Used:** It is used when comparing observed frequencies to expected frequencies, such as in contingency tables or to test if data fits a specific distribution. It is also used in tests involving large sample sizes.

18. What is the Chi-square goodness of fit test, and how is it applied?
>>> **a. Chi-square Goodness of Fit Test:** A statistical test used to determine if a sample data matches an expected distribution.

**b. Application:** It compares observed frequencies in different categories with expected frequencies. If the difference is significant, the null hypothesis (that the data fits the expected distribution) is rejected.

19. What is the F-distribution, and when is it used in hypothesis testing?
>>> **a. F-distribution:** A probability distribution used to compare variances between two or more groups.

**b. When Used:** It is used in tests like ANOVA (Analysis of Variance) to determine if there are significant differences in variances across groups. It is also used in regression analysis to assess the overall fit of the model.

20. What is an ANOVA test, and what are its assumptions?
>>> **a. ANOVA Test:** A statistical test used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

**b. Assumptions:**  
   - Independence of observations.  
   - Normality of the data within each group.  
   - Homogeneity of variances (equal variances across groups).

21. What are the different types of ANOVA tests?
>>> **a. One-way ANOVA:** Compares the means of three or more independent groups based on one factor (e.g., different treatments or conditions).

**b. Two-way ANOVA:** Compares the means of groups based on two factors, considering both their individual effects and any interaction between them.

**c. Repeated Measures ANOVA:** Compares means across three or more related groups (e.g., measurements taken on the same subjects at different times).

22. What is the F-test, and how does it relate to hypothesis testing?
>>> **a. F-test:** A statistical test used to compare two variances or the overall fit of models in regression or ANOVA.

**b. Relation to Hypothesis Testing:** In ANOVA, the F-test checks if there are significant differences between group means. In regression, it tests whether the model explains a significant amount of variance in the dependent variable. The F-statistic is calculated by comparing the variance between groups or models to the variance within groups.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

# Given data
sample_data = [23, 25, 30, 22, 27, 28, 26, 24, 29, 30]  # Sample data
population_mean = 27  # Known population mean
population_std_dev = 3  # Known population standard deviation
sample_size = len(sample_data)

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
sample_data = [98, 102, 105, 97, 101, 99, 103, 95, 100, 104]  # Sample data
population_mean = 100  # Known population mean
population_std_dev = 4  # Known population standard deviation
alpha = 0.05  # Significance level

# Calculate sample mean and sample size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-score for two-tailed test
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Hypothesis test interpretation
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")

# Plotting the Z-distribution with decision regions
x = np.linspace(-4, 4, 1000)  # X values for Z-distribution
y = stats.norm.pdf(x, 0, 1)  # Probability density function (normal distribution)

# Plot the Z-distribution curve
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Highlight the rejection regions (two-tailed test)
plt.fill_between(x, y, where=(x <= -stats.norm.ppf(1-alpha/2)) | (x >= stats.norm.ppf(1-alpha/2)), color='red', alpha=0.5, label="Rejection Region")

# Mark the Z-score on the plot
plt.axvline(z_score, color='green', linestyle='dashed', label=f"Z-Score = {z_score:.2f}")

# Customize the plot
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend(loc="upper left")
plt.grid(True)

# Show the plot
plt.show()
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(population_mean, sample_mean, population_std_dev, sample_size, alpha=0.05, true_population_mean=None):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample_mean) if isinstance(sample_mean, np.ndarray) else sample_mean
    sample_size = len(sample_mean) if isinstance(sample_mean, np.ndarray) else sample_size

    # Step 1: Calculate Z-score for hypothesis testing
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Step 2: Critical Z-value (for a two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Z-value for the rejection region

    # Step 3: Calculate Type 1 Error (α)
    type_1_error = alpha  # This is simply the significance level for a two-tailed test

    # Step 4: Calculate Type 2 Error (β)
    # If we assume an alternative hypothesis true population mean, calculate the Z for that case
    if true_population_mean is not None:
        z_beta = (z_critical * (population_std_dev / np.sqrt(sample_size))) + true_population_mean - population_mean
        beta = stats.norm.cdf(z_beta - (population_mean - true_population_mean))  # Probability of failing to reject H₀ when H₁ is true
    else:
        beta = 0  # If no true population mean is provided, we cannot calculate Type 2 error.

    # Step 5: Print the errors
    print(f"Type 1 Error (α): {type_1_error}")
    print(f"Type 2 Error (β): {beta}")

    # Step 6: Visualization
    # Create the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    # Plot the standard normal distribution
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    # Highlight the rejection regions (Type 1 Error)
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label="Rejection Region (Type 1 Error)")

    # Highlight the region for Type 2 Error (fail to reject H₀ when H₁ is true)
    if true_population_mean is not None:
        z_beta_value = (true_population_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
        plt.fill_between(x, y, where=(x >= z_beta_value), color='green', alpha=0.5, label="Type 2 Error Region (Fail to Reject H₀)")

    # Mark the Z-score
    plt.axvline(z_score, color='green', linestyle='dashed', label=f"Z-Score = {z_score:.2f}")

    # Customize plot
    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper left")
    plt.grid(True)

    # Show plot
    plt.show()

# Example usage:
population_mean = 100
sample_mean = np.random.normal(101, 4, 30)  # Simulating a sample mean for testing
population_std_dev = 4
sample_size = 30
true_population_mean = 102  # Assuming the true mean is 102 for Type 2 Error calculation

# Call the function
visualize_errors(population_mean, sample_mean, population_std_dev, sample_size, alpha=0.05, true_population_mean=true_population_mean)
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample_data (array-like): The sample data.
    confidence_level (float): The confidence level for the interval (default is 0.95).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Calculate the sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(sample_data)

    # Calculate the critical value (z or t value) based on the confidence level
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1

    if sample_size > 30:  # Use Z-distribution for large samples (n > 30)
        critical_value = stats.norm.ppf(1 - alpha / 2)
    else:  # Use T-distribution for small samples
        critical_value = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
sample_data = np.random.normal(100, 15, 50)  # Simulate sample data with mean=100, std=15, n=50
confidence_level = 0.95

# Calculate the confidence interval
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval ({confidence_level*100}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
#10 Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculate the margin of error for a given sample data and confidence level.

    Parameters:
    sample_data (array-like): The sample data.
    confidence_level (float): The confidence level (default is 0.95).

    Returns:
    float: The margin of error for the sample data.
    """
    # Sample size, mean, and standard deviation
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

    # Calculate the critical value (z or t value) based on the confidence level
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1

    if sample_size > 30:  # Use Z-distribution for large samples (n > 30)
        critical_value = stats.norm.ppf(1 - alpha / 2)
    else:  # Use T-distribution for small samples
        critical_value = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std_dev / np.sqrt(sample_size))

    return margin_of_error

# Example usage:
sample_data = np.random.normal(100, 15, 50)  # Simulate sample data with mean=100, std=15, n=50
confidence_level = 0.95

# Calculate the margin of error
margin_of_error = calculate_margin_of_error(sample_data, confidence_level)

print(f"Margin of Error for {confidence_level*100}% Confidence Level: {margin_of_error:.2f}")

```

```python
#11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, false_positive, evidence):
    """
    Perform Bayesian Inference to compute the posterior probability.

    Parameters:
    - prior (float): The prior probability of the hypothesis (e.g., probability of having the disease).
    - likelihood (float): The likelihood of observing the evidence given the hypothesis (e.g., P(Test positive | Disease)).
    - false_positive (float): The probability of observing the evidence given the negation of the hypothesis (e.g., P(Test positive | No disease)).
    - evidence (float): The total probability of the evidence (e.g., P(Test positive)).

    Returns:
    - posterior (float): The posterior probability of the hypothesis given the evidence.
    """
    # Calculate posterior probability using Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Given values for the problem:
prior_disease = 0.01  # Prior probability of having the disease (1%)
likelihood_disease = 0.9  # Likelihood of testing positive if you have the disease (90%)
false_positive_rate = 0.05  # Likelihood of testing positive if you don't have the disease (5%)
test_result_positive = 0.01 * 0.9 + 0.99 * 0.05  # Total probability of testing positive (P(Test positive))

# Apply Bayesian Inference
posterior_disease_given_positive = bayesian_inference(
    prior_disease,
    likelihood_disease,
    false_positive_rate,
    test_result_positive
)

print(f"Posterior Probability of having the disease given a positive test result: {posterior_disease_given_positive:.4f}")
```

```python
#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed_data (numpy array): The observed frequency data (contingency table).

    Returns:
    numpy array: The expected frequencies.
    """
    # Get the total number of observations (sum of all observed values)
    total = np.sum(observed_data)

    # Calculate the row totals and column totals
    row_totals = np.sum(observed_data, axis=1, keepdims=True)
    col_totals = np.sum(observed_data, axis=0, keepdims=True)

    # Calculate the expected frequencies
    expected_data = (row_totals * col_totals) / total

    return expected_data

# Example observed data (Gender vs Beverage Preference)
# Rows: Male, Female
# Columns: Tea, Coffee
observed_data = np.array([[30, 10],  # 30 males prefer tea, 10 males prefer coffee
                          [20, 20]])  # 20 females prefer tea, 20 females prefer coffee

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#16 Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1 (array-like): The first sample data.
    - sample2 (array-like): The second sample data.

    Returns:
    - f_statistic (float): The calculated F-statistic.
    - p_value (float): The p-value from the F-distribution.
    """
    # Calculate the sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate the degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate the F-statistic
    f_statistic = var1 / var2 if var1 >= var2 else var2 / var1  # Larger variance goes in the numerator

    # Calculate the p-value based on the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value

# Simulating two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with mean=50, std=10
sample2 = np.random.normal(loc=50, scale=15, size=100)  # Sample 2 with mean=50, std=15

# Perform the F-test
f_stat, p_val = f_test(sample1, sample2)

# Output the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpret the result
alpha = 0.05
if p_val <= alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference in variances.")
```

```python
#17 Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
import numpy as np
import scipy.stats as stats

def perform_anova(*groups):
    """
    Perform a one-way ANOVA test to compare the means of multiple groups.

    Parameters:
    - groups: Groups of data to compare (as separate arguments).

    Returns:
    - f_statistic (float): The calculated F-statistic.
    - p_value (float): The p-value from the ANOVA test.
    """
    # Perform the ANOVA test
    f_statistic, p_value = stats.f_oneway(*groups)

    return f_statistic, p_value

# Simulating data for three groups
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1 with mean=50, std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Group 2 with mean=55, std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Group 3 with mean=60, std=10

# Perform the ANOVA test
f_stat, p_val = perform_anova(group1, group2, group3)

# Output the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation of results
alpha = 0.05
if p_val <= alpha:
    print("\nReject the null hypothesis: At least one group mean is significantly different.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference in the group means.")
```

```python
#19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st

def check_normality(*groups):
    """
    Perform the Shapiro-Wilk test for normality on each group.

    Parameters:
    - groups: Groups of data to test for normality.

    Returns:
    - results (dict): A dictionary with group names as keys and p-values for normality test as values.
    """
    results = {}
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        results[f'Group {i} Normality Test'] = p_value
    return results

def check_homogeneity_of_variance(*groups):
    """
    Perform Levene's test for equality of variances across groups.

    Parameters:
    - groups: Groups of data to test for homogeneity of variance.

    Returns:
    - p_value (float): The p-value from Levene's test.
    """
    stat, p_value = stats.levene(*groups)
    return p_value

def plot_qq(*groups):
    """
    Plot QQ plots to visually assess the normality of each group.

    Parameters:
    - groups: Groups of data to create QQ plots for.
    """
    for i, group in enumerate(groups, 1):
        plt.figure(figsize=(6, 4))
        st.probplot(group, dist="norm", plot=plt)
        plt.title(f'QQ Plot - Group {i}')
        plt.show()

def check_independence(*groups):
    """
    This function cannot directly test independence, but it can remind the user to ensure
    that data collection meets the independence assumption.

    Parameters:
    - groups: Groups of data.
    """
    print("Ensure that data collection follows random sampling to meet the independence assumption.")
    print("Independence assumption cannot be tested directly in this function.")

def check_anova_assumptions(*groups):
    """
    Check assumptions for ANOVA: normality, homogeneity of variance, and independence.

    Parameters:
    - groups: Groups of data to check assumptions for.

    Returns:
    - normality_results (dict): p-values for the normality test for each group.
    - homogeneity_p_value (float): p-value for the Levene's test.
    """
    # Check normality using Shapiro-Wilk test
    normality_results = check_normality(*groups)

    # Check homogeneity of variance using Levene's test
    homogeneity_p_value = check_homogeneity_of_variance(*groups)

    # Plot QQ plots for visual inspection of normality
    plot_qq(*groups)

    # Check independence assumption
    check_independence(*groups)

    # Print summary
    print("\nNormality Test Results:")
    for group, p_value in normality_results.items():
        print(f"{group}: p-value = {p_value:.4f}")

    print(f"\nLevene's Test for Homogeneity of Variance: p-value = {homogeneity_p_value:.4f}")

    if homogeneity_p_value <= 0.05:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are equal.")

    return normality_results, homogeneity_p_value

# Simulating data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Check the assumptions for ANOVA
normality_results, homogeneity_p_value = check_anova_assumptions(group1, group2, group3)
```

```python
#21 Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(dfn, dfd, x_range=(0, 5), num_points=1000):
    """
    Plot the F-distribution for given numerator (dfn) and denominator (dfd) degrees of freedom.

    Parameters:
    - dfn: degrees of freedom for the numerator (e.g., between-group variance)
    - dfd: degrees of freedom for the denominator (e.g., within-group variance)
    - x_range: the range of x-values for plotting
    - num_points: the number of points for the plot

    """
    # Generate x values for the plot
    x = np.linspace(x_range[0], x_range[1], num_points)

    # Calculate the F-distribution PDF (Probability Density Function)
    pdf_values = f.pdf(x, dfn, dfd)

    # Plot the F-distribution
    plt.plot(x, pdf_values, label=f'F-distribution (dfn={dfn}, dfd={dfd})')
    plt.fill_between(x, 0, pdf_values, alpha=0.2)
    plt.title(f'F-distribution (dfn={dfn}, dfd={dfd})')
    plt.xlabel('F-statistic')
    plt.ylabel('Density')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

# Visualize F-distribution with different degrees of freedom
plot_f_distribution(dfn=2, dfd=10)  # Example 1: F-distribution with dfn=2, dfd=10
plot_f_distribution(dfn=5, dfd=20)  # Example 2: F-distribution with dfn=5, dfd=20
plot_f_distribution(dfn=10, dfd=30) # Example 3: F-distribution with dfn=10, dfd=30
```

```python
#24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data (sample data)
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=50)  # Sample size = 50, population variance = 10^2

# Step 2: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)

# Step 3: State the null and alternative hypotheses
# H₀: The population variance is 100 (σ₀^2 = 100)
# H₁: The population variance is not 100 (σ₀^2 ≠ 100)

sigma_0_squared = 100  # Hypothesized population variance
n = len(sample_data)  # Sample size

# Step 4: Compute the Chi-square test statistic
chi_square_statistic = (n - 1) * sample_variance / sigma_0_squared

# Step 5: Degrees of freedom
df = n - 1

# Step 6: Find the critical values for a two-tailed test at significance level α = 0.05
alpha = 0.05
lower_critical_value = stats.chi2.ppf(alpha / 2, df)  # Lower critical value
upper_critical_value = stats.chi2.ppf(1 - alpha / 2, df)  # Upper critical value

# Step 7: Calculate the p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

# Step 8: Output the results
print(f"Sample variance: {sample_variance:.2f}")
print(f"Chi-square statistic: {chi_square_statistic:.2f}")
print(f"Degrees of freedom: {df}")
print(f"Critical values: lower = {lower_critical_value:.2f}, upper = {upper_critical_value:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 9: Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from 100.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from 100.")
```

```python
#25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

# Step 1: Define the sample data
# Group 1: 100 people, 60 successes
# Group 2: 120 people, 80 successes
n1 = 100  # Sample size of group 1
x1 = 60   # Number of successes in group 1

n2 = 120  # Sample size of group 2
x2 = 80   # Number of successes in group 2

# Step 2: Calculate the sample proportions
p1 = x1 / n1  # Proportion of successes in group 1
p2 = x2 / n2  # Proportion of successes in group 2

# Step 3: Calculate the pooled proportion
p = (x1 + x2) / (n1 + n2)

# Step 4: Calculate the standard error
SE = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Step 5: Calculate the Z-statistic
Z = (p1 - p2) / SE

# Step 6: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

# Step 7: Output the results
print(f"Group 1 Proportion: {p1:.2f}")
print(f"Group 2 Proportion: {p2:.2f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 8: Interpretation of results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

```python
#27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (e.g., dice roll outcomes)
# Suppose we roll a fair 6-sided die 120 times and get the following outcomes
observed_frequencies = np.array([20, 22, 18, 15, 25, 20])  # Observed frequencies for each face of the die

# Step 2: Calculate expected frequencies for a fair 6-sided die (uniform distribution)
total_rolls = np.sum(observed_frequencies)  # Total number of rolls
expected_frequencies = np.full_like(observed_frequencies, total_rolls / 6)  # Each outcome should occur with equal probability

# Step 3: Perform the Chi-square test
chi_square_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# Step 4: Output the results
print(f"Observed frequencies: {observed_frequencies}")
print(f"Expected frequencies: {expected_frequencies}")
print(f"Chi-square statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpretation of results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")

# Step 6: Visualize the observed vs expected frequencies
labels = ['1', '2', '3', '4', '5', '6']  # Labels for the 6 faces of the die

x = np.arange(len(labels))

# Plot observed and expected frequencies
plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, observed_frequencies, width=0.4, label='Observed', color='blue', align='center')
plt.bar(x + 0.2, expected_frequencies, width=0.4, label='Expected', color='orange', align='center')
plt.xticks(x, labels)
plt.xlabel('Dice Faces')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
